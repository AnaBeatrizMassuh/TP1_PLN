```python
import re
import pandas as pd
```

criação do objeto dataframe

```python
df = pd.read_csv("https://raw.githubusercontent.com/viniciusrpb/cic0269_natural_language_processing/main/diario_oficial_pref_sp/dosp_29012016.csv")
df
```

Selecionando linhas e colunas

```python
t = df.iloc[:7,6]
t
```

Título da Publicação

```python
titles = []
for line in t:
  title = re.findall(r"..TITULO..(.*)\s*..TEXTO..", line)
  print(title)
  titles.append(title[0])
```

Número do Processo

```python
processes = []
for line in t:
  process = re.findall(r".*.SEI(.*)..CL..", line)
  if(process == []):
    process.append("NaN")
  print(process)
  processes.append(process[0])
```

Valor (Valor Global): valor real após "R$"

```python
values = []
for line in t:
  value = re.findall(r".*VALOR GLOBAL.*..CL....(..........)", line)
  if(value == []):
    value.append("NaN")
  print(value)
  values.append(value[0])
```

```python
list = []
for i in range(t.size):
  list.append([titles[i], processes[i], values[i]])

tabela = pd.DataFrame(list, columns = ['Título', 'Número do Processo', 'Valor'])

tabela
```
